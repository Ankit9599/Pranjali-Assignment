```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
df = df[df.Ticker.str.match('BANKNIFTY\w+CE.NFO')| df.Ticker.str.match('BANKNIFTY\w+PE.NFO')]
```

```python
df['scrip'] = df['Ticker'].str.extract('([0-9]{5}[A-z]E)')
df['expiry'] = df['Ticker'].str.extract('([0-9]{2}[A-Z]{3}[0-9]{2})')
```

```python
df['expiry'] = pd.to_datetime(df['expiry'])
```

```python
df['date_time'] = df['Date'] + ' ' + df['Time']
df['date_time'] = pd.to_datetime(df['date_time'])
```

```python
df['date_time'] = df['date_time'].dt.strftime('%d/%m/%Y %H:%M')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['days_to_expiry'] = df['expiry']-df['Date']
df['expiry'] = df['expiry'].dt.strftime('%d/%m/%Y')
```

```python
df = df.drop(['Ticker'], axis=1)
df = df.drop(['Date'], axis=1)
df = df.drop(['Time'], axis=1)
df = df.drop(['Volume'], axis=1)
df = df.drop(['Open Interest'], axis=1)
```

```python
df['days_to_expiry'] = df['days_to_expiry'].astype(str)
df.dtypes
```

```python
df.isna().sum()
```

```python
df['dte'] = df['days_to_expiry'].str.extract('([-]?[0-9]{1,2})')
df['dte'] = df['dte'].astype(int)
```

```python
df = df.drop('days_to_expiry', axis = 1)
```

```python
df
```

```python
shift_column = df.pop('scrip')
df.insert(0, 'scrip', shift_column)
shift_column = df.pop('date_time')
df.insert(0, 'date_time', shift_column)
df=df.reset_index(drop=True)
df.set_index('date_time')
```
